```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import genfromtxt
```

```python
# Get suicides for England across all years
df = pd.read_excel( 'suicide2.xls', index_col = 0, usecols = 'A,B')
plt.rcParams["figure.figsize"] = (15,7)
df.plot(kind="line" )
plt.title("Suicides in England")
plt.xlabel("Year")
plt.ylabel("Deaths")
plt.savefig('England.png')
plt.show()
```

```python
# Get suicides for parts of England across all years
df = pd.read_excel( 'suicide2.xls', index_col = 0, usecols = 'A,C,L,W,AF,AP,AY,BK,BN,CH')
plt.rcParams["figure.figsize"] = (15,7)
df.plot(kind="line" )
plt.title("Suicides in Parts of England")
plt.xlabel("Year")
plt.ylabel("Deaths")
plt.legend(bbox_to_anchor=(1, 1), loc='upper left', borderaxespad=0.)
plt.savefig('PartEngland.png')
plt.show()
```

```python
# Box plot for parts of England
df = pd.read_excel( 'suicide2.xls', index_col = 0, usecols = 'A,C,L,W,AF,AP,AY,BK,BN,CH')
plt.rcParams["figure.figsize"] = (15,6)
df.plot(kind="box" )
plt.title("Distribution of suicide")
plt.xlabel("Location")
plt.ylabel("Deaths")
plt.xticks(rotation=30, horizontalalignment="center")
plt.savefig('box.png')
plt.show()
```

```python
# get average for each year and make a piechart
df = pd.read_excel( 'suicide2.xls' ,index_col = 0, usecols = 'A,B')
# df = df.drop('ENGLAND', axis=1)
df = df.mean(axis=1)
plt.rcParams["figure.figsize"] = (15,7)
df.plot(kind="pie",autopct='%1.2f%%' )
plt.title("Average across years")
plt.savefig('Pie.png')
plt.show()
```

```python
# Make time series for males with jobs in Major group
df = pd.read_excel( 'suicidesByOccupation.xlsx', sheet_name = 'major', index_col = 0)
plt.rcParams["figure.figsize"] = (13,9)
df.plot(kind="line")
plt.title("Male-Major")
plt.xlabel("Year")
plt.ylabel("deaths")
plt.legend(bbox_to_anchor=(1,1), loc='upper left', borderaxespad=0.)
plt.savefig('Mmajor.png')
plt.show()
```

```python
# Get averages of each sheet and save them in variables
mmCount = 0
for i in range (1,10):
    df = pd.read_excel( 'suicidesByOccupation.xls', sheet_name = 'major', nrows = 1
                       ,skiprows = ([i for i in range(1,i)]))
    df = df.drop('Year', axis=1)
    mmCount = mmCount + df.mean(axis = 1)
    mmAverage = mmCount/9


mfCount = 0
for i in range (1,10):
    df = pd.read_excel( 'suicidesByOccupation.xlsx', sheet_name = 'majorF', nrows = 1
                       ,skiprows = ([i for i in range(1,i)]))
                       
    df = df.drop('Year', axis=1)
    mfCount = mfCount + df.mean(axis = 1)
    mfAverage = mfCount/9


sbCount = 0
for i in range (1,10):
    df = pd.read_excel( 'suicidesByOccupation.xlsx', sheet_name = 'sub-major', nrows = 1
                       ,skiprows = ([i for i in range(1,i)]))
                       
    df = df.drop('Year', axis=1)
    sbCount = sbCount + df.mean(axis = 1)
    sbAverage = sbCount/9


sbfCount = 0
for i in range (1,10):
    df = pd.read_excel( 'suicidesByOccupation.xlsx', sheet_name = 'sub-majorF', nrows = 1
                       ,skiprows = ([i for i in range(1,i)]))
                       
    df = df.drop('Year', axis=1)
    sbfCount = sbfCount + df.mean(axis = 1)
    sbfAverage = sbfCount/9


mimCount = 0
for i in range (1,10):
    df = pd.read_excel( 'suicidesByOccupation.xlsx', sheet_name = 'minor', nrows = 1
                       ,skiprows = ([i for i in range(1,i)]))     
    df = df.drop('Year', axis=1)
    mimCount = mimCount + df.mean(axis = 1)
    mimAverage = mimCount/9


mimfCount = 0
for i in range (1,10):
    df = pd.read_excel( 'suicidesByOccupation.xlsx', sheet_name = 'minorF', nrows = 1
                       ,skiprows = ([i for i in range(1,i)]))     
    df = df.drop('Year', axis=1)
    mimfCount = mimfCount + df.mean(axis = 1)
    mimfAverage = mimfCount/9


uCount = 0
for i in range (1,10):
    df = pd.read_excel( 'suicidesByOccupation.xlsx', sheet_name = 'unit', nrows = 1
                       ,skiprows = ([i for i in range(1,i)]))     
    df = df.drop('Year', axis=1)
    uCount = uCount + df.mean(axis = 1)
    uAverage = uCount/9


ufCount = 0
for i in range (1,10):
    df = pd.read_excel( 'suicidesByOccupation.xlsx', sheet_name = 'unitF', nrows = 1
                       ,skiprows = ([i for i in range(1,i)]))     
    df = df.drop('Year', axis=1)
    ufCount = ufCount + df.mean(axis = 1)
    ufAverage = ufCount/9
```

```python
#Change the datatype of the variables so you could plot them
mmAv = float(mmAverage) 
sbAv = float(sbAverage)
mimAv = float(mimAverage)
uAv = float(uAverage)
mfAv = float(mfAverage)
sbfAv = float(sbfAverage)
mimfAv = float(mimfAverage)
ufAv = float(ufAverage)
```

```python
# Make stacked bar chart of each sheet
plotdata = pd.DataFrame({
    "Male":[mmAv, sbAv, mimAv, uAv],
    "Female":[mfAv, sbfAv, mimfAv, ufAv]
    }, 
    index=["Major", "Sub-Major", "minor", "Unit"]
)

plotdata.plot(kind="bar", stacked=True)
plt.title("Mean number of suicides across group")
plt.xlabel("Group")
plt.ylabel("Suicides")
plt.xticks(rotation=0, horizontalalignment="center")
plt.savefig('stack.png')
```
